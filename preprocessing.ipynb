```python
#Preprocessing dataset 
import pandas as pd 
import string
```

```python
df= pd.read_csv('F:\Edu\covidQaBot\dataset\Covid-QA.csv')
df.head()
```

```python
dataset=df.drop(labels=['answer_html', 'region','city','lang','last_update'], axis=1)
dataset.head()
```

```python
#Check null values
dataset['answer'].isnull().values.any()


```

```python
#Remove whitspaces
dataset['answer'] = dataset['answer'].str.strip()
```

```python
#Check duplication
bool_series = dataset['answer'].duplicated()
bool_series.values
```

```python
#drop duplication
dataset=dataset.drop_duplicates()
dataset.shape
```

```python
bool_series = dataset.duplicated()
bool_series.values
```

```python
#Remove Punctuation from Answers
string.punctuation
```

```python
#Remove Punctuation Function
def remove_punctuation(txt):
    txt_nopunt="".join([c for c in txt if c not in string.punctuation])
    return txt_nopunt
```

```python
dataset['ans_clean']=dataset['answer'].apply(lambda x: remove_punctuation(x))
dataset.head()
```

```python
#remove old column Answer and rename no punctuation answer
dataset=dataset.drop(labels='answer', axis=1)
dataset.rename(columns = {'ans_clean':'answer'}, inplace = True)
dataset.head()
```

```python
#lowercase dataset 
dataset['answer']=dataset['answer'].apply(str.lower)
dataset
```
